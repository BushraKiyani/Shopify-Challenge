## b. What is the last name of the employee with the most orders?

SELECT LastName <br>
FROM (SELECT LastName, MAX(total) <br>
FROM(SELECT e.LastName, count(o.EmployeeID)total FROM Employees AS e, Orders AS o <br>
WHERE e.EmployeeID = o.EmployeeID <br>
GROUP BY o.EmployeeID, e.LastName));

### Employee's Last Name: Peacock

## c. What product was ordered the most by customers in Germany?

SELECT ProductName <br>
FROM (SELECT ProductName, MAX(MostOrdered) <br>
FROM (SELECT p.ProductName, count(*) MostOrdered FROM Customers AS c, Orders As o, OrderDetails As od, Products As p <br>
WHERE c.Country = 'Germany'and c.CustomerID = o.CustomerID and o.OrderID = od.OrderID and od.ProductID = p.ProductID <br>
GROUP BY p.ProductName));

### Product Name: Gorgonzola Telino
